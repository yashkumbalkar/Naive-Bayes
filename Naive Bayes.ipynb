# **Text Cleaning**

```python
# Taking only 10K rows

df = df.sample(10000)
```

```python
df.shape
```

```python
df['sentiment'].value_counts()
```

```python
df.info()
```

```python
df['sentiment'] = df['sentiment'].map({'positive':1 , 'negative':0})
```

```python
df.head()
```

```python
# Function to clean html tags
def clean_html(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)
```

```python
df['review'] = df['review'].apply(clean_html)
```

```python
# converting everything to lower
def convert_lower(text):
    return text.lower()
```

```python
df['review'] = df['review'].apply(convert_lower)
```

```python
# function to remove special characters
def remove_special(text):

   x = ''

   for i in text:
     if i.isalnum():
       x = x + i
     else:
       x = x + ' '
   return x
```

```python
df['review'] = df['review'].apply(remove_special)
```

```python
# Remove the stop words
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
def remove_stopwords(text):

    x = []

    for i in text.split():
      if i not in stopwords.words('english'):
        x.append(i)
    return x
```

```python
df['review'] = df['review'].apply(remove_stopwords)
```

```python
df
```

```python
# Perform stemming

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem_words(text):

    y = []

    for i in text:
        y.append(ps.stem(i))
    return y
```

```python
df['review'] = df['review'].apply(stem_words)
```

```python
# Join back

def join_back(list_input):
    return " ".join(list_input)
```

```python
df['review'] = df['review'].apply(join_back)
```

```python
df
```

# **Vectorization**

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500)
```

```python
X = cv.fit_transform(df['review']).toarray()
```

```python
X.shape
```

```python
y = df['sentiment']
y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape , X_test.shape
```

# **Naive Bayes**

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB

clf1 = GaussianNB()
clf2 = MultinomialNB()
clf3 = BernoulliNB()
```

```python
clf1.fit(X_train,y_train)
clf2.fit(X_train,y_train)
clf3.fit(X_train,y_train)
```

```python
y_pred1 = clf1.predict(X_test)
y_pred2 = clf2.predict(X_test)
y_pred3 = clf3.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print("Gaussian Naive Bayes Accuracy" , accuracy_score(y_test,y_pred1))
print("Multinomial Naive Bayes Accuracy" , accuracy_score(y_test,y_pred2))
print("Bernaulli Naive Bayes Accuracy" , accuracy_score(y_test,y_pred3))
```
